```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        # area = x*y
        # base = start-end
        area = []
        for i in range(0,len(height)):
            for j in range(1,len(height)):
                # base = i1-i2
                base = i*j
                iheight = min(height[i],height[j])
                area.append(base*iheight)
                
        print(max(area))
        return max(area)
```
